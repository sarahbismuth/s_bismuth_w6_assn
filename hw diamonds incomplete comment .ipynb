## Q1 What proportion of diamonds are between .30 and 1.08 carats?
##### Ans : Applied the condition on dataframe column carat greater than .30 and less than equal to 1.08. This will give TRUE for the meeting criteria and False for the non matching records. 
##### And, then putting the results of this condtion in variable. Passing the same variable in pandas dataframe to get the resultset with TRUE value only. 


```python
cond1_30_8_carat = (df['carat'] > .30) & (df['carat'] <= 1.08) 
#len will give the records satisfying the given condition
len(df[cond1_30_8_carat])
```

From this command we can see that the proportion of diamonds between 0.30 and 1.08 carats is 37680.
